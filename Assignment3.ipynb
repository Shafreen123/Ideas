```python
#Question8 text preprocessing
import pandas as pd # importing necessary modules
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
nltk.download('punkt') # downloady explicitly as its not accessing implicitely, and therefore throwing error
nltk.download('stopwords')
nltk.download('wordnet')
df = pd.read_csv("/content/titanic/train.csv") # loading the dataset
print("Missing values in 'Embarked' column:", df['Embarked'].isnull().sum()) # removing null values from embarked column as its contain some NaN
def preprocess_text(text): # text preprocessing function , separetly doing each function was quite messy
    text = str(text) # converting to string
    text = text.translate(str.maketrans('', '', string.punctuation + string.digits)) #  removing punctuation and digits
    text = text.lower() # converting test in lower case
    tokens = word_tokenize(text) # tokenizes the text into individual words
    stop_words = set(stopwords.words('english')) # removes commen used words
    tokens = [token for token in tokens if token not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens] # lemmatizes each token to its base form using NLTK's WordNet lemmatizer
    processed_text = ' '.join(tokens) # Joins the lemmatized tokens back into a single string.
    return processed_text
df['Embarked'].fillna('', inplace=True) # filling any missing value in embarked
df['processed_text'] = df['Embarked'].apply(preprocess_text)
print(df['processed_text'].head())
```
